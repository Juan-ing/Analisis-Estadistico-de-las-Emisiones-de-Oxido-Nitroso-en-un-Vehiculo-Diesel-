```python
##importación de librerías
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats

# Configuración estética
sns.set(style="whitegrid")

# Mostrar todas las columnas
pd.set_option('display.max_columns', None)
```

```python
# Cargar el dataset
df = pd.read_excel("../data/oxido nitroso.xls")
#Mostrar todos los datos
df.head(20)
```

```python
#Información general del DataFrame
df.info()
```

```python
#Estadísticas descriptivas
df.describe()
```

```python
#Revisión de valores nulos
df.isnull().sum()
```

```python
# Renombrar columnas con nombres y unidades
df.columns = [
    "NOx (ppm)",        # Óxido nitroso
    "Humidity (%)",     # Humedad relativa
    "Temp (°C)",        # Temperatura
    "Pressure (inHg)"   # Presión atmosférica
]

# Mostrar primeras filas
df.head()
```

```python
# Visualización de histogramas para cada variable
df.hist(figsize=(10,8), bins=15)

plt.tight_layout()
plt.show()
```

```python
#Diagramas de caja para detectar valores atípicos
plt.figure(figsize=(10,6))

for i, col in enumerate(df.columns, 1):
    plt.subplot(2,2,i)
    sns.boxplot(y=df[col])

plt.tight_layout()
plt.show()
```

```python
# Matriz de correlación
corr = df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Matriz de Correlación")
plt.show()
```

```python
# Gráficos de dispersión para cada par de variables
sns.pairplot(df)

plt.show()
```

```python
# Conclusiones Exploratorias

# No se evidencian valores faltantes en la base de datos.
# Las variables presentan comportamientos aproximadamente continuos.
# Se observa una posible relación lineal entre las emisiones de óxido nitroso y algunas variables ambientales.
# La variable con mayor correlación aparente será analizada posteriormente mediante regresión lineal simple.
# No se identifican valores atípicos extremos que comprometan inicialmente el análisis.
```
